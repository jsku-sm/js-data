### 종속변수 (금융이해력) 기초 통계량 해석

위 표는 금융이해력을 구성하는 각 문항(Q02~Q14)에 대한 기초 통계량을 보여줍니다. 주요 통계량은 다음과 같습니다:

*   **count (개수)**: 각 문항에 응답한 표본의 수를 나타냅니다. 이전에 확인한 결측치를 제외한 실제 응답자 수입니다.
*   **mean (평균)**: 각 문항에 대한 평균 점수입니다. 문항별 평균을 통해 응답자들이 해당 금융 지식이나 행위에 대해 얼마나 이해하고 실천하는지 전반적인 경향을 파악할 수 있습니다.
*   **std (표준편차)**: 데이터가 평균으로부터 얼마나 퍼져 있는지를 나타냅니다. 표준편차가 클수록 응답자 간의 편차가 크다는 것을 의미합니다.
*   **min (최소값)**: 각 문항의 최소 점수입니다.
*   **max (최대값)**: 각 문항의 최대 점수입니다.
*   **25%, 50%, 75% (사분위수)**: 데이터 분포의 특성을 파악할 수 있는 지표로, 각각 하위 25%, 50%(중앙값), 75%에 해당하는 값입니다.

**해석 시 고려사항:**

*   **문항별 난이도 및 응답 경향**: 문항별 평균 점수를 비교하여 특정 금융 지식이나 행위 문항에서 더 높은/낮은 이해도를 보이는지 파악할 수 있습니다.
*   **점수 척도 확인**: 각 문항이 어떤 척도(예: 1점~4점, 1점~5점 등)로 구성되어 있는지 코드북을 통해 확인하면 평균 점수의 의미를 더 정확하게 해석할 수 있습니다. (예: 4점 만점에서 평균 3점은 비교적 높은 이해도를 의미할 수 있습니다.)
*   **표준편차 비교**: 표준편차가 큰 문항은 응답자 간의 이해도나 실천 정도의 차이가 크다는 것을 의미하므로, 추가적인 탐색이 필요할 수 있습니다.

이 통계량은 개별 문항에 대한 기초적인 이해도를 제공하며, 이후 금융이해력 종합점수를 생성하기 위한 전 단계로 활용됩니다.

### 종속변수 (금융이해력 각 문항) 상자 그림 해석

위 상자 그림은 금융이해력을 구성하는 각 문항(Q02~Q14)의 점수 분포를 시각적으로 보여줍니다. 상자 그림을 통해 다음을 파악할 수 있습니다:

*   **중앙값 (Median)**: 상자 안의 가로선은 각 문항의 중앙값을 나타냅니다. 이는 응답자의 절반이 이 점수보다 낮거나 같고, 나머지 절반은 이 점수보다 높거나 같다는 것을 의미합니다.
*   **사분위 범위 (Interquartile Range, IQR)**: 상자의 높이는 25번째 백분위수(Q1)부터 75번째 백분위수(Q3)까지의 범위를 나타냅니다. 이 범위 안에 전체 데이터의 50%가 포함됩니다. 상자의 길이가 길수록 응답자 간 점수 편차가 크다는 것을 의미합니다.
*   **수염 (Whiskers)**: 상자 위아래로 뻗어 있는 수염은 Q1 - 1.5 * IQR과 Q3 + 1.5 * IQR 사이의 데이터 범위를 나타냅니다. 이 수염의 길이는 데이터의 분포 범위를 보여줍니다.
*   **이상치 (Outliers)**: 수염 바깥에 점으로 표시된 값들은 이상치(극단값)로, 다른 데이터와 동떨어진 경향을 보이는 값들입니다. 이는 특정 응답자들이 해당 문항에서 매우 높거나 낮은 점수를 받았음을 시사할 수 있습니다.

**해석 시 고려사항:**

*   **문항별 난이도 비교**: 각 문항의 중앙값이나 상자의 위치를 비교하여 어떤 금융 지식이나 행위 문항에서 응답자들이 더 높은/낮은 점수를 받는지 파악할 수 있습니다.
*   **점수 분포의 대칭성**: 중앙선이 상자 중앙에 가까울수록 분포가 대칭적이라고 볼 수 있습니다. 한쪽으로 치우쳐 있다면 왜도(skewness)가 있음을 의미합니다.
*   **이상치의 유무 및 개수**: 이상치가 많이 나타나는 문항은 특정 소수 그룹의 점수가 전체 분포와 크게 다르다는 것을 의미할 수 있으며, 이는 추가적인 분석이 필요할 수 있습니다.

이 시각화는 개별 금융이해력 문항의 전반적인 분포와 응답자들의 경향을 이해하는 데 도움을 줍니다.

### 독립변수 히스토그램 (분포 확인)

```python
import matplotlib.pyplot as plt
import seaborn as sns

indep_var_cols = [
    'Q26', # 부모 금융교육 수준
    'Q27', # 학교 금융교육 경험
    'Q33'  # 금융 웰빙
]

variable_titles = {
    'Q26': '부모 금융교육 수준',
    'Q27': '학교 금융교육 경험',
    'Q33': '금융 웰빙'
}

plt.figure(figsize=(18, 5))

for i, col in enumerate(indep_var_cols):
    plt.subplot(1, 3, i + 1) # 1행 3열의 서브플롯
    # Ensure bins are appropriate for categorical/ordinal data, assuming scores are integers
    if df[col].min() is not None and df[col].max() is not None:
        sns.histplot(df[col].dropna(), kde=True, bins=range(int(df[col].min()), int(df[col].max()) + 2))
    else:
        sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'히스토그램: {variable_titles[col]}', fontsize=14)
    plt.xlabel(variable_titles[col], fontsize=12)
    plt.ylabel('빈도', fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

독립변수 히스토그램 해석
위 세 히스토그램은 연구의 주요 독립변수인 '부모 금융교육 수준 (Q26)', '학교 금융교육 경험 (Q27)', '금융 웰빙 (Q33)'의 분포를 보여줍니다.

부모 금융교육 수준 (Q26):

히스토그램을 보면 점수가 낮은 쪽(좌측)에 많은 응답자가 분포하고, 점수가 높아질수록 빈도가 줄어드는 오른쪽으로 치우친(right-skewed) 분포를 보입니다. 이는 중·고등학생들이 부모로부터 받는 금융교육 수준이 비교적 낮거나, 특정 교육 항목에 집중되어 있을 가능성을 시사합니다. 전반적으로 부모의 금융교육 참여가 활발하지 않거나, 학생들의 인식이 낮은 경향이 있을 수 있습니다.
학교 금융교육 경험 (Q27):

이 변수 또한 낮은 점수대에 많은 응답자가 몰려 있는 오른쪽으로 치우친 분포를 나타냅니다. 이는 학교에서 금융교육을 경험한 학생의 수가 상대적으로 적거나, 교육의 강도나 범위가 제한적일 수 있음을 보여줍니다. 즉, 학교 금융교육의 보편성이 아직은 낮다고 해석할 수 있습니다.
금융 웰빙 (Q33):

'금융 웰빙' 변수는 비교적 고르게 분포되어 있거나, 중간 점수대에 가장 많은 응답자가 분포하는 경향을 보일 수 있습니다. (정확한 분포는 히스토그램의 형태를 다시 한번 확인해야 하지만, 일반적으로 웰빙 척도는 정규분포에 가까운 형태를 띠는 경우가 많습니다.) 이는 학생들이 느끼는 금융 불안감이나 경제적 여유에 대한 인식이 다양하게 분포되어 있음을 의미하며, 극도로 높거나 낮은 수준보다는 중간 정도의 금융 웰빙을 느끼는 학생들이 많다고 볼 수 있습니다.
종합적으로 보면, 부모와 학교를 통한 외부적인 금융교육의 경험은 낮은 수준에 집중되어 있는 반면, 개인의 내면적인 '금융 웰빙' 인식은 좀 더 분산된 경향을 보입니다. 이러한 분포는 향후 다중회귀분석 시 각 변수가 금융이해력에 미치는 독립적인 영향력을 분석하는 데 중요한 기초 자료가 될 것입니다. 특히, 금융교육 변수들의 치우친 분포는 교육 효과 분석 시 주의 깊게 다루어져야 할 부분입니다.

### 독립변수 결측치 처리 (행 제거)

```python
print(f"원본 데이터프레임 크기: {df.shape}")

# 독립변수 컬럼 리스트
indep_vars_for_na = ['Q26', 'Q27', 'Q33']

# 해당 독립변수 컬럼에 결측치가 하나라도 있는 행 제거
df_cleaned_indep = df.dropna(subset=indep_vars_for_na).copy()

print(f"독립변수 결측치 제거 후 데이터프레임 크기: {df_cleaned_indep.shape}")
print("제거 후 데이터프레임의 첫 5행을 표시합니다.")
display(df_cleaned_indep.head())
```

## Step 2. 더미화 및 분석용 데이타프레임 생성

### 분석용 데이터프레임 (df_analysis) 재구성 (통제변수 제외)

사용자 요청에 따라, 종속변수와 독립변수만을 포함하고 통제변수는 제외하여 `df_analysis` 데이터프레임을 다시 구성합니다. '부모 금융교육 수준 (Q26)', '금융 웰빙 (Q33)'과 같은 연속형 독립변수와 '금융이해력 관련 문항 (Q02~Q14)'을 포함하며, '학교 금융교육 경험 (Q27)'은 더미 변수로 변환됩니다.

```python
# 종속변수 (금융이해력 관련 문항)
dep_var_cols = [
    'Q02', 'Q03', 'Q04', 'Q05', 'Q06', 'Q07', 'Q08', # 금융지식
    'Q09', 'Q10', 'Q11', 'Q12', 'Q13', 'Q14'  # 금융행위
]

# 연속형 독립변수
continuous_indep_var_cols = [
    'Q26', # 부모 금융교육 수준 (합산점수)
    'Q33'  # 금융 웰빙 (합산점수)
]

# 더미 변수화할 독립변수 (학교 금융교육 경험만 해당, 통제변수 제외)
categorical_indep_vars_for_dummy = [
    'Q27' # 학교 금융교육 경험
]

# 1. 범주형 독립변수를 더미 변수로 변환 (drop_first=True로 다중공선성 방지)
df_indep_dummies = pd.get_dummies(df_cleaned_indep[categorical_indep_vars_for_dummy], columns=categorical_indep_vars_for_dummy, drop_first=True)

# 2. 분석에 사용할 종속변수, 연속형 독립변수, 더미 변수화된 독립변수들을 df_analysis에 결합
df_analysis = pd.concat([
    df_cleaned_indep[dep_var_cols], # 종속변수
    df_cleaned_indep[continuous_indep_var_cols], # 연속형 독립변수
    df_indep_dummies # 더미 변수화된 독립변수
], axis=1)

print("### df_analysis 데이터프레임 재구성 완료 (통제변수 제외) ###")
print("df_analysis의 첫 5행을 표시합니다.")
display(df_analysis.head())
print(f"df_analysis 데이터프레임 크기: {df_analysis.shape}")
```

### VIF (Variance Inflation Factor) 계산을 통한 다중공선성 확인

다중회귀분석을 수행하기 전에 독립변수들 간의 다중공선성 문제를 확인하는 것이 중요합니다. VIF 값은 독립변수가 다른 독립변수들에 의해 얼마나 설명되는지를 나타내며, 일반적으로 VIF가 10 이상이면 다중공선성 문제가 심각하다고 판단합니다. 각 독립변수에 대해 VIF 값을 계산하여 출력합니다.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd
import statsmodels.api as sm

# 종속변수 (금융이해력 관련 문항) 정의
dep_var_cols = [
    'Q02', 'Q03', 'Q04', 'Q05', 'Q06', 'Q07', 'Q08', # 금융지식
    'Q09', 'Q10', 'Q11', 'Q12', 'Q13', 'Q14'  # 금융행위
]

# 금융이해력 종합점수 계산
df_analysis['financial_literacy_score'] = df_analysis[dep_var_cols].sum(axis=1)

# financial_literacy_score에 결측치가 있는 행 제거하여 df_analysis_final 생성
df_analysis_final = df_analysis.dropna(subset=['financial_literacy_score']).copy()

# 독립변수 설정 (통제변수 제외)
X_cols = [
    'Q26',         # 부모 금융교육 수준
    'Q33',         # 금융 웰빙
    'Q27_2.0'      # 학교 금융교육 경험 (더미 변수)
]
X_for_vif = df_analysis_final[X_cols]

# 상수항 추가 (VIF 계산을 위해 필요)
X_for_vif = sm.add_constant(X_for_vif)

# VIF 계산
vif_data = pd.DataFrame()
vif_data["variable"] = X_for_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_for_vif.values, i) for i in range(X_for_vif.shape[1])]

print("### 독립변수 VIF 값 ###")
display(vif_data.round(2))
```

## Step 4. 다중선형회귀

### 다중선형회귀분석 실행 및 결과 해석

다중선형회귀분석을 통해 정의된 독립변수들이 종속변수인 '금융이해력 종합점수'에 미치는 영향을 분석합니다. OLS(Ordinary Least Squares) 모델을 사용하여 회귀분석을 수행하고, 그 결과를 바탕으로 각 변수의 통계적 유의미성과 영향력을 해석합니다.

```python
import statsmodels.api as sm

# 종속변수 (Y): 금융이해력 종합점수
Y = df_analysis_final['financial_literacy_score']

# 독립변수 (X): Q26, Q33, Q27_2.0
X = df_analysis_final[['Q26', 'Q33', 'Q27_2.0']]

# 회귀 모델에 상수항 추가 (절편을 계산하기 위해 필요)
X = sm.add_constant(X)

# OLS (Ordinary Least Squares) 모델 적합
model = sm.OLS(Y, X)
results = model.fit()

# 회귀분석 결과 출력
print("### 다중선형회귀분석 결과 ###")
display(results.summary())
```

### 다중선형회귀분석 결과 해석

위 출력된 회귀분석 결과(`OLS Regression Results`)는 독립변수들이 금융이해력 종합점수에 미치는 영향을 자세히 보여줍니다. 주요 지표들은 다음과 같습니다:

*   **R-squared**: 모델이 종속변수의 분산을 얼마나 잘 설명하는지를 나타냅니다. 0부터 1까지의 값을 가지며, 1에 가까울수록 설명력이 높습니다. 조정된 R-squared(`Adj. R-squared`)는 변수의 개수를 고려하여 R-squared를 조정한 값으로, 모델의 일반화 능력을 평가하는 데 더 유용합니다.

*   **F-statistic & Prob (F-statistic)**: 모델 전체의 통계적 유의성을 나타냅니다. `Prob (F-statistic)` 값이 0.05보다 작으면 모델이 통계적으로 유의미하다고 판단하며, 독립변수 중 적어도 하나는 종속변수에 유의미한 영향을 미친다는 것을 의미합니다.

*   **Coefficients (coef)**: 각 독립변수가 한 단위 증가할 때 종속변수가 얼마나 변하는지를 나타내는 회귀계수입니다. 양수이면 양의 상관관계, 음수이면 음의 상관관계를 의미합니다.

*   **Std.Err. (표준 오차)**: 회귀계수 추정치의 정밀도를 나타냅니다. 표준 오차가 작을수록 추정치가 더 정확합니다.

*   **t**: 회귀계수가 0과 유의미하게 다른지 검정하는 t-통계량입니다.

*   **P>|t| (P-value)**: 각 독립변수의 회귀계수가 통계적으로 유의미한지 판단하는 유의확률입니다. 일반적으로 이 값이 0.05보다 작으면 해당 변수가 종속변수에 통계적으로 유의미한 영향을 미친다고 해석합니다.

*   **[0.025, 0.975] (신뢰구간)**: 회귀계수에 대한 95% 신뢰구간입니다. 이 구간 안에 0이 포함되지 않으면 해당 변수는 통계적으로 유의미하다고 볼 수 있습니다.

**해석 가이드라인:**

1.  **모델의 전반적인 유의성 확인**: `Prob (F-statistic)` 값을 통해 모델이 전반적으로 금융이해력에 유의미한 영향을 미치는지 확인합니다.
2.  **독립변수의 설명력**: `R-squared` 값을 통해 독립변수들이 금융이해력의 변화를 얼마나 잘 설명하는지 파악합니다.
3.  **개별 독립변수의 유의성과 영향**: 각 독립변수의 `P>|t|` 값을 보고 0.05보다 작은지 확인하여 통계적 유의성을 판단합니다. 유의미한 변수에 대해서는 `coef` 값을 통해 금융이해력에 미치는 영향의 방향(양수/음수)과 크기를 해석합니다.

이 결과를 바탕으로 부모 금융교육, 학교 금융교육, 금융 웰빙이 청소년의 금융이해력에 어떤 영향을 주는지 구체적으로 논의할 수 있습니다.

    *   **부모 금융교육 수준 (Q26)**: 회귀계수(coef)는 0.3343, p-값은 0.000으로 **통계적으로 매우 유의미하며 긍정적인 영향**을 미칩니다. 부모로부터의 금융교육 수준이 높을수록 청소년의 금융이해력 종합점수가 증가하는 경향을 보였습니다.

### 5. 결론 및 향후 연구 제언

본 연구의 다중선형회귀분석 결과, **청소년의 금융이해력은 부모 금융교육 수준과 학교 금융교육 경험에 의해 긍정적인 영향을 받는 것으로 나타났습니다.** 이는 가정과 학교에서의 금융교육이 청소년의 금융지식 함양 및 금융행위 실천에 중요한 역할을 한다는 것을 지지합니다. 반면, 금융 웰빙은 이 모델에서 직접적인 유의미한 영향을 보이지 않았습니다.

**주요 한계점 및 향후 연구 제언:**

1.  **낮은 설명력**: R-squared 값이 낮다는 것은 청소년 금융이해력에 영향을 미치는 다른 중요한 요인들이 많으며, 이 모델에서 설명하지 못하는 부분이 크다는 것을 의미합니다. 학년, 성별, 부모학력 등 통제변수를 포함한 추가 분석이 필요합니다.
2.  **매개효과 분석**: 연구 주제의 핵심인 **'금융 웰빙의 매개효과'**는 본 다중선형회귀분석으로는 직접적으로 검증되지 않았습니다. 금융 웰빙이 부모/학교 금융교육과 금융이해력 사이에서 어떤 역할을 하는지 파악하기 위해서는 구조방정식 모델링(Structural Equation Modeling)과 같은 매개효과 분석 방법론을 적용하는 것이 필수적입니다. 이는 금융 웰빙이 금융교육의 효과를 강화하거나 약화시키는 경로를 밝히는 데 기여할 것입니다.
